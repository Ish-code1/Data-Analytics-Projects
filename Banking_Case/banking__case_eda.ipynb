```python
bins = [0,100000,300000,float('inf')]
labels = ['Low','Med','High']

df['Income Band'] = pd.cut(df['Estimated Income'], bins =bins , labels = labels, right = False)
```

```python
df['Income Band'].value_counts()
```

```python
#Distibutions in Catagorical Data
```

```python
cat_cols = df[['BRId', 'GenderId', 'Amount of Credit Cards', 'Nationality', 'Occupation', 'Fee Structure',
              'Loyalty Classification', 'Properties Owned','Risk Weighting','IAId']]

for col in cat_cols:
  print(f"Value Counts for '{col}': ")
  display(df[col].value_counts())
```


**Univariate Analysis**

```python
for i , predictor in enumerate(df[['BRId', 'GenderId', 'Amount of Credit Cards', 'Nationality', 'Occupation', 'Fee Structure',
              'Loyalty Classification', 'Properties Owned','Risk Weighting','IAId']]):
    plt.figure(i)
    sns.countplot(data=df,x=predictor)
```

**Bivariate Analysis**

```python
for i , predictor in enumerate(df[['BRId', 'GenderId', 'Amount of Credit Cards', 'Nationality', 'Occupation', 'Fee Structure',
              'Loyalty Classification', 'Properties Owned','Risk Weighting','IAId']]):
    plt.figure(i)
    sns.countplot(data=df,x=predictor,hue='GenderId')
```

```python
for i , predictor in enumerate(df[['BRId', 'GenderId', 'Amount of Credit Cards', 'Nationality', 'Occupation', 'Fee Structure',
              'Loyalty Classification', 'Properties Owned','Risk Weighting','IAId']]):
    plt.figure(i)
    sns.countplot(data=df,x=predictor,hue='Nationality')
```

```python
#Histplot for value counts for different Ocuupation

for col in cat_cols:
  if col == "Occupation":
    continue
  plt.figure(figsize=(8,4))
  sns.histplot(df[col])
  plt.title('Histogram of Occupation Count')
  plt.xlabel(col)
  plt.ylabel('Count')
  plt.show()
```

```python
numerical_cols =['Estimated Income', 'Superannuation Savings','Credit Card Balance','Bank Loans', 'Bank Deposits','Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

plt.figure(figsize=(15,10))
for i,col in enumerate(numerical_cols):
  plt.subplot(4,3,i+1)
  sns.histplot(df[col],kde =True)
  plt.title(col)
  plt.show()
```

##Heat Maps

```python
numerical_cols =['Estimated Income', 'Superannuation Savings','Credit Card Balance','Bank Loans', 'Bank Deposits','Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix,annot=True,fmt='.2f',cmap='coolwarm',linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
